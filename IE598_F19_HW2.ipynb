```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:/Users/Administrator/Desktop/Treasury Squeeze test - DS1.csv')

y = df['squeeze'].values
X = df.drop('squeeze', axis = 1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

k_range = range(1,26)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k, p = 2, metric = 'minkowski')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))
print(scores)

plt.plot(k_range, scores, c = 'black')
plt.show()

dt = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, random_state = 1)
dt.fit(X_train, y_train)
y_pred_tree = dt.predict(X_test)
tree_score = accuracy_score(y_test, y_pred_tree)
print(tree_score)

print("My name is Jiaqi Su")
print("My NetID is: jiaqisu2")
print("I hereby certify that I have read the University policy on Academic Integrity and that I am not in violation.")
```
